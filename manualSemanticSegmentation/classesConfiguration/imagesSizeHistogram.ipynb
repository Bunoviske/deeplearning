```python
from fastai.basics import *
from fastai.callback.all import *
from fastai.vision.all import *
from PIL import Image


path = 'C:/Users/bruno/Documents/diabetesLearning/foodDatabase/Dataset - Comidas Brasileiras TCC v1/dataset/done/'
fnames = get_image_files(path)
len(fnames)
```

```python
biggestImageArea = 0
smallestImageArea = 5000*5000
biggestImage = None
smallestImage = None

areas = []
sizes = []

for file in fnames:
    img = Image.open(file)
    width = int(img.size[0])
    height = int(img.size[1])
    area = width*height
    areas.append(area)
    if area > biggestImageArea:
        biggestImageArea = area
        biggestImage = (width, height)
    if area < smallestImageArea:
        smallestImageArea = area
        smallestImage = (width, height)
    if area < 95230:
        sizes.append((width, height))
    
print("Smallest image", smallestImage)
print("Biggest image", biggestImage)
```

```python
#maxRange = max(areas)
maxRange = 1500000
hist = plt.hist(areas, bins=20, range=(0, maxRange))
hist
```

```python
histDict = {}

for cont, area in zip(hist[0], hist[1]):
    dim = int(math.sqrt(area))
    name = "{}x{}".format(dim, dim)
    histDict[name] = cont
```

```python
histDict
plt.bar(*zip(*histDict.items()))
plt.xticks(rotation=45)

plt.show()
```
